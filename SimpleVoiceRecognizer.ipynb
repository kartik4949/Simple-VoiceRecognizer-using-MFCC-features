```python
import librosa    
import os
import numpy as np
from pathlib import Path    
  
```

```python
MFCC = []
label = []
```

```python


pathlist = Path("Data/").glob('**/*.wav')
for path in pathlist:
   
    path_in_str = str(path)
    X, sample_rate = librosa.load(path_in_str, res_type='kaiser_fast') 
    mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T,axis=0)
    MFCC.append(mfccs)
    
    if "jackson" in path_in_str:
                    
                    label.append(1)
    else:
    
        label.append(0)
    
    
    
```

```python

from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(80, input_dim=40, activation='relu'))
model.add(Dense(80, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python

X=np.asarray(MFCC)
y=np.asarray(label)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((-1,1))
scaler.fit(X)
X=scaler.transform(X)

model.fit(X, y, epochs=150, batch_size=10)
```

```python
x = []
n = []
pathlist = Path("Test/").glob('**/*.wav')
for path in pathlist:
   
    path_in_str = str(path)
    b, sample_rate = librosa.load(path_in_str, res_type='kaiser_fast') 
    mfccs = np.mean(librosa.feature.mfcc(y=b, sr=sample_rate, n_mfcc=40).T,axis=0)
    x.append(mfccs)
    
    if "jackson" in path_in_str:
                    
                    n.append(1)
    else:
    
        n.append(0)

        scaler.fit(x)
x=scaler.transform(x)
inpu = np.asarray(x)
predictions = model.predict(inpu)

predictions
```
